Import data

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import warnings
warnings.filterwarnings('ignore')

features = pd.read_csv('features.csv')
stores = pd.read_csv('stores.csv')
train_raw = pd.read_csv('train.csv')
test_raw = pd.read_csv('test.csv')
```

Set date format

```python
train_raw['Date'] = pd.to_datetime(train_raw['Date'])
test_raw['Date'] = pd.to_datetime(test_raw['Date'])
features['Date'] = pd.to_datetime(features['Date'])
```

Merge Train and Test

```python
all_raw = pd.concat([train_raw, test_raw], 0)
```

Merge features and train

```python
merged_1 = pd.merge(all_raw, stores, how = 'left', on = 'Store')
merged = pd.merge(merged_1, features, how = 'left', on = ['Store', 'Date'])

merged['Week'] = merged['Date'].dt.week

merged.tail()
```

```python
def dummies(df, col_name):
    dum = pd.get_dummies(df[col_name], prefix=col_name)
    df = pd.concat([df, dum], axis = 1)
    df.drop(col_name, axis = 1, inplace = True)
    return(df)

one_hot = dummies(merged, 'Type')
one_hot = dummies(one_hot, 'Week')

one_hot.loc[one_hot['MarkDown1'].isnull(), 'MarkDown1'] = 0
one_hot.loc[one_hot['MarkDown2'].isnull(), 'MarkDown2'] = 0
one_hot.loc[one_hot['MarkDown3'].isnull(), 'MarkDown3'] = 0
one_hot.loc[one_hot['MarkDown4'].isnull(), 'MarkDown4'] = 0
one_hot.loc[one_hot['MarkDown5'].isnull(), 'MarkDown5'] = 0
one_hot['IsHoliday'] = one_hot['IsHoliday_x'] * 1
one_hot.drop(['IsHoliday_x', 'IsHoliday_y', 'CPI', 'Unemployment'], 1, inplace = True)

```
